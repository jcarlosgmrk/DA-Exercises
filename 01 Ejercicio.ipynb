```python
import numpy as np
import pandas as pd

iris = pd.read_csv("Iris.csv")
iris
```

```python
# Determine en una sola tabla la media, desviación estándar y la mediana (percentil 50) para el ancho del pétalo por cada tipo de flor 
# para aquellos registros cuyo ancho del sépalo es mayor o igual a 3.5 y el largo del sépalo es mayor que 5.

iris_filtrar = iris[(iris['sepal.length'] > 5) & (iris['sepal.width'] >= 3.5)]
ancho_petalo_flor = iris_filtrar.groupby('variety')['petal.width'].agg(['mean','std','median'])
ancho_petalo_flor
```

```python
# Calcule el número de registros que se utilizaron para obtener los resultados del problema 1.
registros = iris_filtrar.shape[0]
registros
```

```python
# Agregue una nueva columna al DataFrame generado al problema 1, que sea resultado de estandarizar los valores del ancho del pétalo para cada registro.
# Denomine a dicha columna “petal.width.st” y calcúlela de la siguiente manera:

media_petal = iris_filtrar['petal.width'].mean()
std_petal = iris_filtrar['petal.width'].std()

iris_filtrar.loc[:,'petal.width.st'] = (iris_filtrar['petal.width'] - media_petal) / std_petal

iris_filtrar
```

```python
# Mantenga en el nuevo DataFrame del problema 3 las columnas “variety” y “petal.width.st”. Renómbrelas como “Variedad” y “Ancho Est. Pétalo” 

iris_filtrar = iris_filtrar.rename(columns = {'variety':'Variedad','petal.width.st':'Ancho Est. Pétalo'})
iris_filtrar
```

```python
# Exporte el resultado del DataFrame obtenido en el problema 4 a un archivo denominado “Iris2.csv”. 

iris_filtrar.to_csv("Iris2.csv")
```
