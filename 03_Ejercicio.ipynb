```python
import pandas as pd
df = pd.read_csv("SYB66_246_202310_Population Growth, Fertility and Mortality Indicators.csv")
df
```

```python
# Mediante el uso de grÃ¡ficos de su preferencia, muestre la relaciÃ³n que guardan la tasa de mortalidad infantil (infant mortality for both sexes) 
# con la esperanza de vida para ambos sexos (life expectancy at birth for both sexes) para el aÃ±o 2020.

import seaborn as sns
import matplotlib.pyplot as plt

filtrar_1 = df[(df["Region/Country/Area"] == 1) & (df["Year"] == 2020)]
filtrar_filas_1 = filtrar_1[filtrar_1["Series"].isin(["Infant mortality for both sexes (per 1,000 live births)", "Life expectancy at birth for both sexes (years)"])]

filtrar_filas_1.loc[:, "Value"] = pd.to_numeric(filtrar_filas_1["Value"])

plt.figure(figsize=(10, 10))
sns.barplot(data=filtrar_filas_1, x = "Series", y = "Value", alpha=0.7, hue = "Series", palette=["red", "blue"])
plt.xlabel("AÃ±o 2020")
plt.ylabel("Valor")
plt.title("GrÃ¡fico del total de todas las regiones")
plt.xticks(rotation=45)
plt.savefig("TareaM10_1.jpg", dpi=300, bbox_inches='tight')
plt.show()
```

```python
df_year = df[df['Year'] == 2020]

df_year["Value"] = pd.to_numeric(df_year["Value"], errors="coerce")
numeric_vars = df_year.select_dtypes(include='number')

corr = numeric_vars.corr()
corr_notna = ~corr.isna().all()
corr = corr.loc[corr_notna, corr_notna]

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, vmax=1, vmin=-1)
plt.title('Matriz de CorrelaciÃ³n para el AÃ±o 2020 (Kendall)')
plt.savefig("TareaM10_5.jpg",dpi=300, bbox_inches='tight')
plt.show()
```
