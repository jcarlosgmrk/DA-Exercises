```python
from openpyxl.utils.cell import get_column_letter
from openpyxl.utils.dataframe import dataframe_to_rows
nuevo_archivo = "nuevo_archivo.xlsx"
columnas = [3, 4, 5, 6] 

wb_nuevo = openpyxl.Workbook()
ws_nuevo = wb_nuevo.active

for col_index in columnas:
        columna_letra = get_column_letter(col_index)
        for row_index, row in enumerate(sheet.iter_rows(min_col=col_index, max_col=col_index), start=1):
            for cell in row:
                ws_nuevo[columna_letra + str(row_index)].value = cell.value
    
# Guardar el nuevo libro de trabajo
wb_nuevo.save(nuevo_archivo)
```

## JSON

```python
import json
from pprint import pprint
with open('season1.json', 'r') as json_file:
    temporada_json = json.load(json_file)
pprint(temporada_json)
```

```python
# Hacemos un print de todo el contenido del capitulo 2 para comprobar que todo funcione
for num, texto in temporada_json['Game Of Thrones S01E02 The Kingsroad.srt'].items():
    print(f"{num}: {texto}")
```

```python
# Imprimimos las primeras 10 filas del capitulo 1
for clave in sorted(temporada_json['Game Of Thrones S01E01 Winter Is Coming.srt'])[:10]:
    print(f"{clave}: {temporada_json['Game Of Thrones S01E01 Winter Is Coming.srt'][clave]}")
```

```python
# Imprimimos las 10 ultimas filas del capitulo 1
for clave in sorted(temporada_json['Game Of Thrones S01E01 Winter Is Coming.srt'])[-10:]:
    print(f"{clave}: {temporada_json['Game Of Thrones S01E01 Winter Is Coming.srt'][clave]}")
```

```python
# Obtenemos el primer dialogo de cada capitulo y lo almacenamos
dialogos_1 = {}
for num, texto in temporada_json.items():
    if '1' in texto:
        dialogos_1[num] = texto['1']

        #print(f"{num}: {texto}")
pprint(dialogos_1)
```

```python
nuevo_json = 'primer_dialogo.json'
with open(nuevo_json, 'w') as json_file:
    json.dump(dialogos_1, json_file, indent=4)
```
