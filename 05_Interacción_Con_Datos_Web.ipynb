```python
# Uso de librería BeautifulSoup y requests. Parseo de información.
import requests
from bs4 import BeautifulSoup
from pprint import pprint

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0',
}

url = 'https://es.aliexpress.com/w/wholesale-audifonos.html?spm=a2g0o.home.search.0'
data = requests.get(url, headers)

title_size = []

html = BeautifulSoup(data.text, 'lxml')
articles = html.select('.multi--outWrapper--SeJ8lrF')

for article in articles:
    try:
        print('--------------------------------')
        title = article.select('.multi--title--G7dOCj3')[0].get_text()

        print(title)
    except Exception as e:
        #print(' ')
        raise e
```
