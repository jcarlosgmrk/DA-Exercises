```python
# Añadir una columna “Years Playing”, que calcule el año actual menos la columna “Joined”
df['Years Playing'] = ''
```

```python
df.dtypes
```

```python
# Cambiamos el tipo de dato de object a int64
df['Years Playing'] = pd.to_numeric(df['Years Playing'], errors='coerce')
df['Years Playing'].fillna(0, inplace=True)
df['Years Playing'] = df['Years Playing'].astype('int64')
```

```python
df.dtypes
```

```python
df['Years Playing'] = 2024 - df['Joined']
```

```python
# Buscar y mostrar a todos los jugadores de Mexico
df[['ID', 'Name', 'Age','Club','Nationality']].loc[df['Nationality'] == 'Mexico']
```

```python
# Ordenar y mostrar los datos por la columna ReleaseClause (sueldo)
df[['ID', 'Name', 'Age','Club','Release Clause','Nationality']].loc[df['Nationality'] == 'Mexico'].sort_values(by='Release Clause', ascending = True)
```

```python
# Generar un nuevo dataset que contenga el año (joined) y el número de jugadores (groupby)
df_temp = df.groupby('Joined').size().rename('numero_jugadores').reset_index()
df_temp
```

```python
# Generar un gráfico que contenga, por año, el número de jugadores
import matplotlib.pyplot as plt
ax = df_temp.plot.bar(x='Joined', y='numero_jugadores', rot=90)
plt.show()
```
