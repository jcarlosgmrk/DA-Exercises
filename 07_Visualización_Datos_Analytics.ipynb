```python
import pandas as pd

df = pd.read_csv("fifa_eda.csv")
df
```

```python
#Elemento 1: Obtener matriz de correlación y dejarla como heat map 
import numpy as np
import plotly.express as px
df_numerico = df.select_dtypes(include=[np.number])
fig = px.imshow(df_numerico.corr(), text_auto=True, title='HeadMap de la matriz de correlación', color_continuous_scale='RdBu_r')
fig.show()
```

```python
#Elemento 2: Gráfico de Correlación que responda – “Cuál es la relación entre la edad y el overall?” – explicar la salida 
fig = px.imshow(df_numerico.corr(), text_auto=True, title='HeadMap de la matriz de correlación', color_continuous_scale='RdBu_r')
fig.show()
```

```python
df_conteo = df.groupby("Club").count()['Name']
```

```python
#Elemento 3: Generar un gráfico de barras por club que indique el número de jugadores 
fig = px.bar(df_conteo, barmode = "group", title = "Numero de jugadores por club")
fig.show()
```

```python
df.groupby("Club").count()['Name'].sort_values(ascending=False)[0:10]
```

```python
#Elemento 4: Hacer un gráfico multipanel que indique la relación altura (height) vs skill moves, siendo la variable del panel si es zurdo o derecho
fig = px.strip(df, x="Skill Moves", y="Height", facet_col="Preferred Foot")
fig.show()
```

```python
df_conteo_2 = df.groupby("Potential").count()['Name'].reset_index()
df_conteo_2
```

```python
#Elemento 5: Un gráfico que crea importante para mostrar a los potenciales cracks
#Jugadores con más de 85, son candidatos.

fig = px.bar(df_conteo_2[df_conteo_2["Potential"] >84], x= "Potential", y= "Name", text_auto=True, color = "Potential")
fig.update_layout(xaxis=dict(tickmode='linear'))
fig.show()
```
