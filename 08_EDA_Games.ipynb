## ❖Insights: 2 Gráficos adicionales interesantes

```python
# Se hace la correlación al DataFrame para encontrar la relación entre las variables.
df_numerico = df.select_dtypes(include=[np.number])
corr_df = corr = df_numerico.corr(method="pearson")
df_lt = corr_df.where(np.tril(np.ones(corr_df.shape)).astype(bool))
hmap = sns.heatmap(df_lt, cmap="Reds", annot=True)
# El dato interesante aquí es que nos damos cuenta como las ventas globales van muy relacionadas a las ventas que se tienen en NA y EU, osease, la mayoría de las ventas provienen de esas regiones. De igual forma obviamente las ventas de JP y otras ventas tienen obvia relación con las ventas globales.
```

```python
import plotly.express as px
# En este descubrimiento, obtenemos la plataforma que más juegos vendio en su ciclo de vida y se compara con todas las demás plataformas, dando como resultado que la PS2 fue la plataforma que más juegos vendio.
fig = px.pie(df, names = "Platform", values = "Global_Sales", title = "Ventas globales por plataforma", width=800, height= 600)
fig.update_traces(textinfo='label+percent', textposition='inside', showlegend=False)
fig.show()
```

## ❖Crecimiento:

```python
# Qué género ha tenido el mayor y menor crecimiento
years_data = [2017, 2016, 2015,2014]
df_4year = df[df['Year'].isin(years_data)]

fig = px.histogram(df_4year, x="Genre", title="Crecimiento del genero 2014-2017", nbins=20)
fig.show()
# Como se observa el genero que más crecio fue Action y el que menor crecimiento tuvo es Puzzle.
```

```python
df_4year.sample(5)
```

```python
# Qué plataforma ha tenido el mayor y menor crecimiento
fig = px.histogram(df_4year, x="Platform", title="Crecimiento de la plataforma 2014-2017", nbins=20)
fig.show()
# La plataforma que tuvo mayor crecimiento es la PS4, y la que tuvo menor crecimiento es la DS, se puede inferir que ya había terminado su ciclo de vida como plataforma.
```

## ❖2 elementos adicionales que muestren oportunidades de mercado

```python
df_4year.groupby("Platform").sum()['Global_Sales'].sort_values(ascending=False)[0:10]
```

```python
df_4year.groupby("Platform").sum()['Global_Sales'].sort_values(ascending=False)[0:10].plot.bar()
```

* La PS4 tiene un mercado grande, lo que significa que puede haber buenas oportunidades si se enfocan en esta plataforma. Sin embargo, también puede ser un mercado saturado, por lo que habrá mucha competencia.
* XOne y 3Ds tienen potencial como plataformas, pueden tener menos competencia que con PS4, ofreciendo oportunidades para expandirse sin enfrentarse a una alta competencia.

```python
df_4year.groupby("Genre").sum()['Global_Sales'].sort_values(ascending=False)[0:10].plot.pie()
```

* Los generos de Action y Shooter son altamente populares y tienen ventas considerables. Sin embargo, es un mercado saturado, para entrar en estos generos se debe innovar o diferenciarse del resto para destacar.
* Sports es un genero que destaca, son juegos que también tienen demanda. Hay oportunidades en este género.
* Role-Playing sugiere que que los juegos de rol son populares entre los jugadores, donde también hay oportunidades de crecimiento con este genero.
