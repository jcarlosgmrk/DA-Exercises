```python
import numpy as np
import pandas as pd

df = pd.read_csv("supermarket_sales.csv")
df
```

```python
unit_price_np = df[["Unit price"]].to_numpy()
quantity_np = df[["Quantity"]].to_numpy()
```

```python
# Importar la librerÃ­a NumPy.  Generar estadÃ­stica descriptiva bÃ¡sica en las columnas unit_price y quantity: CÃ¡lculo de la media, mediana, moda
media_up = unit_price_np.mean()
media_q = quantity_np.mean()
print(f"La media del precio unitario es = {media_up:.2f}")
print(f"La media de las cantitades es = {media_q:.2f}")
print("------------------------------------------")

mediana_up = np.median(unit_price_np)
mediana_q = np.median(quantity_np)
print(f"La mediana del precio unitario es = {mediana_up:.2f}")
print(f"La mediana de las cantitades es = {mediana_q:.2f}")
print("------------------------------------------")

moda_up = df["Unit price"].mode()[0]
moda_q = df["Quantity"].mode()[0]
print(f"La moda del precio unitario es = {moda_up:.2f}")
print(f"La moda de las cantitades es = {moda_q:.2f}")

```

```python
# Obtener el ticket promedio: por ciudad
resultado = df.groupby("City")["Total"].mean().reset_index()
resultado.style.format({"Total": "{:.2f}"})
```

```python
# Obtener el ticket promedio: por product line y ciudad
resultado = df.groupby(["City", "Product line"])["Total"].mean().reset_index()
resultado.style.format({"Total": "{:.2f}"})
```

```python
# Obtener el ticket promedio: por gÃ©nero
resultado = df.groupby("Gender")["Total"].mean().reset_index()
resultado.style.format({"Total": "{:.2f}"})
```

```python
# Obtener el ticket promedio: por gÃ©nero y product line
resultado = df.groupby(["Gender", "Product line"])["Total"].mean().reset_index()
resultado.style.format({"Total": "{:.2f}"})
```

```python
# Usar el mÃ©todo rank para generar top 5 de: ventas por ciudad
df["Rank"] = df.groupby("City")["Total"].rank(ascending= False, method = "first")

resultado = df[df["Rank"] <= 5][["Rank", "City", "Total"]].sort_values(by=["City", "Rank"]).reset_index(drop=True)
resultado.style.format({"Rank": "{:.0f}", "Total": "{:.2f}"})
```

```python
# Usar el mÃ©todo rank para generar top 5 de: ventas por member
df["Rank"] = df.groupby("Customer type")["Total"].rank(ascending= False, method = "first")
resultado = df[df["Rank"] <= 5][["Rank", "Customer type", "Total"]].sort_values(by=["Customer type", "Rank"]).reset_index(drop=True)[0:5]
resultado.style.format({"Rank": "{:.0f}", "Total": "{:.2f}"})
```

```python
# Usar el mÃ©todo rank para generar top 5 de: ventas por payment.  Obtener ademÃ¡s el % de aporte de cada categorÃ­a. 
df["Rank"] = df.groupby("Payment")["Total"].rank(ascending= False, method = "first")
ventas_totales = df.groupby("Payment")["Total"].transform('sum')
df["Aporte %"] = (df["Total"] / ventas_totales) * 100
resultado = df[df["Rank"] <= 5][["Rank", "Payment", "Total", "Aporte %"]].sort_values(by=["Payment", "Rank"]).reset_index(drop=True)
resultado.style.format({"Rank": "{:.0f}", "Total": "{:.2f}", "Aporte %": "{:.2f}%"})
```

```python
# Obtener ademÃ¡s el % de aporte de ciudad
total_categoria = df.groupby("City")["Total"].sum().reset_index().rename(columns={"Total": "Ventas por ciudad"})
total = df["Total"].sum()
total_categoria["Aporte"] = (total_categoria["Ventas por ciudad"] / total) * 100
total_categoria["Ventas por ciudad"] = total_categoria["Ventas por ciudad"].map('{:.2f}'.format)
total_categoria["Aporte"] = total_categoria["Aporte"].map('{:.2f}%'.format)

total_categoria
```

```python
# Obtener ademÃ¡s el % de aporte de Genero
total_categoria = df.groupby("Gender")["Total"].sum().reset_index().rename(columns={"Total": "Ventas por genero"})
total = df["Total"].sum()
total_categoria["Aporte"] = (total_categoria["Ventas por genero"] / total) * 100
total_categoria["Ventas por genero"] = total_categoria["Ventas por genero"].map('{:.2f}'.format)
total_categoria["Aporte"] = total_categoria["Aporte"].map('{:.2f}%'.format)

total_categoria
```

```python
# Obtener ademÃ¡s el % de aporte de Customer
total_categoria = df.groupby("Customer type")["Total"].sum().reset_index().rename(columns={"Total": "Ventas por customer"})
total = df["Total"].sum()
total_categoria["Aporte"] = (total_categoria["Ventas por customer"] / total) * 100
total_categoria["Ventas por customer"] = total_categoria["Ventas por customer"].map('{:.2f}'.format)
total_categoria["Aporte"] = total_categoria["Aporte"].map('{:.2f}%'.format)

total_categoria
```
