## Reportes por grupos : estado civil, educaci√≥n, trabajo ‚Äì cu√°ntos registros

```python
# Contar registros por estado civil
marital_counts = df["marital"].value_counts()
print("Conteo por estado civil:")
print(marital_counts)

# Contar registros por educaci√≥n
education_counts = df["education"].value_counts()
print("\nConteo por educaci√≥n:")
print(education_counts)

# Contar registros por trabajo
job_counts = df["job"].value_counts()
print("\nConteo por trabajo:")
print(job_counts)
```

## An√°lisis univariado y bivariado basado en la predicci√≥n (y/n)

```python
# Columnas con valores "yes" y "no"
yes_no_columns = ['default', 'housing', 'loan', 'y']

# Convertir "yes" a 1 y "no" a 0 usando pd.factorize
for col in yes_no_columns:
    df[col], _ = pd.factorize(df[col])
    if df[col].iloc[0] == 1: 
        df[col] = df[col].map({1: 1, 0: 0})
    else:
        df[col] = df[col].map({1: 0, 0: 1})

# Verificar las conversiones
print(df[yes_no_columns].head())
```

```python
#Analisis univariado
cols_num = ["default", "housing", "loan", "y"]


fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 5))
for i, column in enumerate(cols_num):
    sns.histplot(df[column], ax=axes[i], kde=False)
```

## Dividir dataset entre training y test (70:30) 

```python
df2 = df.copy()
```

```python
from sklearn.model_selection import train_test_split

X = df2.drop("y", axis=1)
y=df2["y"]

X = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.30, random_state=1)
```

```python
y_test.head(3)
```

```python
print("{0:0.2f}% data is in training set".format((len(X_train)/len(df2.index))* 100))
print("{0:0.2f}% data is in test set".format((len(X_test)/len(df2.index))* 100))
```
