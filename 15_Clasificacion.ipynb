## Insights

* Debemos traducir los campos "object" a tipo númerico para que la regresión lineal los tome directamente.
* EmployeeID tiene un valor diferente en cada registro, por lo que no sirve para el modelo y se debe eliminar.
* Las columnas EmployeeCount, Over18 y StandardHours tienen solamente un valor, también se tienen que eliminar.
* NumCompaniesWorked y TotalWorkingYears tienen valores nulos en sus registros, tenemos que imputar datos.


```python
# Empezamos con la eliminación de las columnas que no son necesarias para poder seguir haciendo los analisis pertinentes. (Parte del Feature Engineering)
df.drop('EmployeeID', axis=1, inplace=True)
df.drop('EmployeeCount', axis=1, inplace=True)
df.drop('Over18', axis=1, inplace=True)
df.drop('StandardHours', axis=1, inplace=True)
```

```python
# Imputación de datos
# Remplazar con el promedio general
# Calcula el promedio de Income
num_comp_worked_mean = int(df['NumCompaniesWorked'].mean())
num_total_working_years_mean = int(df['TotalWorkingYears'].mean())

# Imputa el promedio a todos los registros definidos con NaN
df['NumCompaniesWorked'] = df['NumCompaniesWorked'].fillna(num_comp_worked_mean)
df['TotalWorkingYears'] = df['TotalWorkingYears'].fillna(num_total_working_years_mean)
```

## Feature Engineering

```python
# Genera una copia del df original
df2 = df.copy()
```

```python
df2.sample(10)
```

```python
# Balanceo de clases
# Obtiene los distintos valores
print(df2['EducationField'].value_counts())
print('')

print(df2['BusinessTravel'].value_counts())
print('')

print(df2['Department'].value_counts())
print('')
```

```python
# Traducimos los campos object a númerico
# Cambia el tipo de columna a Attrition (1/0) y otras
df2['Attrition'] = df2['Attrition'].replace({'Yes':1, 'No':0}, regex=True)
```

```python
# Category codes - Automático
df2.Gender = df2.Gender.astype('category').cat.codes
df2.EducationField = df2.EducationField.astype('category').cat.codes
df2.BusinessTravel = df2.BusinessTravel.astype('category').cat.codes
df2.Department = df2.Department.astype('category').cat.codes
df2.MaritalStatus = df2.MaritalStatus.astype('category').cat.codes
df2.JobRole = df2.JobRole.astype('category').cat.codes
```

```python
df2.sample(3)
```

```python
df2.info()
```

```python
# Estandarizar
# Por ejemplo Income, que tiene un orden de magnitud más grande que el resto

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df2['MonthlyIncome_scale'] = StandardScaler().fit_transform(df2[['MonthlyIncome']])
```

```python
# Eliminar la variable original "MonthlyIncome"
df2.drop('MonthlyIncome', axis=1, inplace=True)
```

```python
#  Se selecciona las columnas de tipo numérico
cols_num = df2.columns[[np.issubdtype(dt, np.number) for dt in df2.dtypes]]

fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(18,16))
for i, column in enumerate(cols_num):
    sns.histplot(df2[column], ax=axes[i//5, i%5], kde=False)
```

## Data Split

```python
from sklearn.model_selection import train_test_split

X = df2.drop('Attrition', axis=1) # El array X (atributos) no va a contener la respuesta (num columnas x m)
y = df2['Attrition'] #Clase a predecir
#Se divide a X y Y en un ratio 70:30

# Convert categorical variables to dummy variables
# X = pd.get_dummies(X, drop_frist=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
X.head(5)
```

```python
y.head(10)
```

```python
print(df2['Attrition'].value_counts())
print(df2['Attrition'].value_counts(normalize=True))
```

```python
# Aplicación de modelo de regresión logística a la variable Attrition
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

# Adapta el modelo a los X_train y y_train (entrenamiento)
model = LogisticRegression(solver='liblinear', class_weight='balanced')

model.fit(X_train, y_train)

y_predict = model.predict(X_test)
```

```python
# Valor original y predicción
z = X_test.copy()
z['Attrition real'] = y_test
z['Attrition prediccion'] = y_predict
z[['Attrition real', 'Attrition prediccion']].head(20)
```

```python
# Genera los coeficientes de la ecuación para validar la importancia de las variables
coef_df = pd.DataFrame(model.coef_)
coef_df.columns = X.columns
# Incluye el Intercept
coef_df['Intercept'] = model.intercept_
coef_df.T
# Un coeficiente positivo indica que a medida que la variable aumenta, la probabilidad de la clase positiva también aumenta.
# Un coeficiente negativo indica que a medida que la variable aumenta, la probabilidad de la clase positiva disminuye.
# La magnitud del coeficiente indica la fuerza del impacto de la variable en la predicción.
```

```python
# Visualización de los resultados del modelo de clasificación
# Genera la matriz de confusion
cm = metrics.confusion_matrix(y_test, y_predict, labels=[1,0])

df_cm = pd.DataFrame(cm, index = [i for i in['1', '0']],
                     columns = [i for i in ['Prediccion 1', 'Prediccion 0']])
plt.figure(figsize=(7,4))
sns.heatmap(df_cm, annot=True, cmap='Accent',fmt='d').set_title("Matriz de Confusión")
```

```python
# Calculo manual de Falsos Positivos, Falsos Negativos, Verdaderos Positivo y Verdaderos Negativos
TP = df_cm.iloc[0][0]
FP = df_cm.iloc[0][1]
FN = df_cm.iloc[1][0]
TN = df_cm.iloc[1][1]
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, classification_report, confusion_matrix

y_pred = model.predict(X_test)
model_score = accuracy_score(y_test, y_pred)
recall_score = recall_score(y_test, y_pred)
precision_score = precision_score(y_test, y_pred)
f1_score = f1_score(y_test, y_pred)

print('Medidas del Modelo')
print('--------------------------------')
print('Accuracy = {0:0.2f}'.format(model_score))
print('Precision = {0:0.2f}'.format(precision_score))
print('Recall = {0:0.2f}'.format(recall_score))
print('F1 Score = {0:0.2f}'.format(f1_score))
print('Roc Auc Score = {0:0.2f}'.format(roc_auc_score(y_test,y_pred)))
```
