```python
import pandas as pd
import numpy as np
import psycopg2

df = pd.read_csv('Analista de Datos M49 Tweets (1).csv')

# Se sacan los 1000 valores y que lleven la palabra clave
df = df[df['text'].str.contains('flight', case=False, na=False)].sample(1000)

# Limpiar la columna negativereason_gold para manejar valores no numericos y evitar errores
def convert_to_float(value):
    try:
        return float(value)
    except ValueError:
        return np.nan 

df['negativereason_gold'] = df['negativereason_gold'].apply(convert_to_float)


# Convertir DataFrame a una lista de tuplas para executemany
data_to_insert = df.apply(lambda row: (
    row['tweet_id'], row['airline_sentiment'], row['airline_sentiment_confidence'], row['negativereason'],
    row['negativereason_confidence'], row['airline'], row['airline_sentiment_gold'], row['name'],
    row['negativereason_gold'], row['retweet_count'], row['text'], row['tweet_coord'], row['tweet_created'],
    row['tweet_location'], row['user_timezone']
), axis=1).tolist()

# Conexión
conexion = psycopg2.connect(
    host="localhost",
    port="5432",
    database="postgres",
    user="postgres",
    password=""
)

cursor = conexion.cursor()

# Crear la tabla si no existe
cursor.execute('''
    CREATE TABLE IF NOT EXISTS "EBAC"."tweets_importados" (
        tweet_id BIGINT,
        airline_sentiment VARCHAR(50),
        airline_sentiment_confidence FLOAT,
        negativereason VARCHAR(255),
        negativereason_confidence FLOAT,
        airline VARCHAR(50),
        airline_sentiment_gold VARCHAR(50),
        name VARCHAR(100),
        negativereason_gold FLOAT,
        retweet_count INT,
        text TEXT,
        tweet_coord VARCHAR(100),
        tweet_created TIMESTAMP,
        tweet_location VARCHAR(255),
        user_timezone VARCHAR(255)
    );
''')

# Consulta que inserta la tabla
insert_query = '''
    INSERT INTO "EBAC"."tweets_importados" (
        tweet_id, airline_sentiment, airline_sentiment_confidence, negativereason,
        negativereason_confidence, airline, airline_sentiment_gold, name, 
        negativereason_gold, retweet_count, text, tweet_coord, tweet_created,
        tweet_location, user_timezone
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
'''

# Executemany inserta los datos seleccionados
cursor.executemany(insert_query, data_to_insert)
conexion.commit()

# Cerrar la conexión
cursor.close()
conexion.close()

print("1000 datos importados exitosamente.")
```

```python
# Conexión
conexion = psycopg2.connect(
    host="localhost",
    port="5432",
    database="postgres",
    user="postgres",
    password=""
)

cursor = conexion.cursor()

# Ejecuta consulta
cursor.execute('SELECT * FROM "EBAC"."tweets_importados"')
# Verifica funcione
resultados = cursor.fetchall()
if resultados:
    df = pd.DataFrame(resultados, columns=[desc[0] for desc in cursor.description])
    print(df)
else:
    print("La consulta no retornó ningún resultado.")

# Cerrar la conexión
cursor.close()
conexion.close()
```

```python
# Filtrar por sentimiento negativo y aerolinea
aerolinea = 'Virgin America'
filtered_df = df[(df['airline_sentiment'].str.lower() == 'negative') & (df['airline'].str.contains(aerolinea, case=False, na=False))]
print(f"Total de tweets negativos en la aerolinea {aerolinea}: {filtered_df.shape[0]}")
```

```python
# Filtrar por sentimiento positivo y número de confianza
confidence = 0.6 
filtered_df = df[(df['airline_sentiment'].str.lower() == 'positive') &
                 (df['airline_sentiment_confidence'] > confidence)]
print(f"Total de tweets positivos con más de {confidence} de confianza: {filtered_df.shape[0]}")
```
